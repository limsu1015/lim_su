#### 모험가들의 레벨업 도전

[배경 이야기]

당신은 마법사, 전사, 도적, 힐러 네 명의 모험가를 이끄는 리더입니다. 

각 캐릭터는 힘, 민첩, 지능, 체력이라는 능력치를 가지고 있습니다. 이 능력치들을 이용해 다양한 도전을 해결하고 사냥을 통해 경험치를 얻어 레벨업을 해야 합니다.

[동료 캐릭터와 능력치 정보]

마법사: 힘(10), 민첩(20), 지능(80), 체력(50)

전사: 힘(80), 민첩(30), 지능(20), 체력(100)

도적: 힘(40), 민첩(80), 지능(30), 체력(60)

힐러: 힘(20), 민첩(40), 지능(70), 체력(70)

[문제 1: 첫 번째 사냥]

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다. 

마법을 사용하려면 지능이 50 이상이어야 합니다. 마법사의 지능이 50 이상인지 확인하세요.

```python
magic_str, magic_dex, magic_int, magic_hp = 10, 20, 80, 50
war_str, war_dex, war_int, war_hp = 80, 30, 20, 100
thief_str, thief_dex, thief_int, thief_hp = 40, 80, 30, 60
heal_str, heal_dex, heal_int, heal_hp = 20, 40, 70, 70


magic_skill = magic_int >= 50

print(magic_skill)
```

[문제 2: 두 번째 사냥]

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다. 전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
war_str, war_dex, war_int, war_hp = 80, 30, 20, 100
war_skill = war_str >= 60 and war_hp >= 80

print(war_skill)
```

[문제 3: 세 번째 사냥]

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다. 도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
thief_str, thief_dex, thief_int, thief_hp = 40, 80, 30, 60

thief_skill = thief_dex >= 70 or thief_str >= 50

print(thief_skill)
```

[문제 4: 네 번째 사냥]

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다. 힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
heal_str, heal_dex, heal_int, heal_hp = 20, 40, 70, 70
heal_skill = heal_int >= 60 and heal_hp >= 50

print(heal_skill)
```

[문제 5: 레벨업]

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다. 각 캐릭터의 경험치는 다음과 같습니다

[경험치]

마법사: 120

전사: 150

도적: 110

힐러: 90

각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요

```python
magic_exp, war_exp, thief_exp, heal_exp = 120, 150, 110, 90
마법사, 전사, 도적, 힐러 = magic_exp >= 100, war_exp >= 100, thief_exp >= 100, heal_exp>= 100
print(마법사, 전사, 도적, 힐러)
```
